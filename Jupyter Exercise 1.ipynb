```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
GDPs = [[1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010], 
        [1.015, 1.33, 2.29, 3.26, 4.951, 6.759, 9.366, 13.131, 15.599]]
```

```python
plt.figure()
plt.plot(*GDPs, 'r--', *GDPs, 'r.', markersize=10)
plt.title('Real US GDP (in trillions)')
plt.xlabel('Year')
plt.ylabel('Trillions of $')
plt.show()
```

While the GDP curve is not exactly linear, the trend seems roughly linear, so we can probably model the relationship between Year and GDP with a line. By inspection, it looks like the line would pass through (1930, 0) and (2010, 16), which indicates that the line would be $$y = (x - 1930) / 5$$


```python
xline = [i for i in range(1930, 2011)]
yline = [(i - 1930)*2/10 for i in xline]
plt.figure()
plt.plot(*GDPs, 'r--', *GDPs, 'r.', xline, yline, 'b:', markersize=10)
plt.title('Real US GDP (in trillions)')
plt.xlabel('Year')
plt.ylabel('Trillions of $')
plt.show()
```
